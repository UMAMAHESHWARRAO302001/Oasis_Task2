```python
ue.columns=['State','Date','Frequency','Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate', 'Area']
```

```python
import datetime as dt
import calendar

ue['Date'] = pd.to_datetime(ue['Date'], dayfirst=True)
ue['month_int'] =  ue['Date'].dt.month
ue['month'] =  ue['month_int'].apply(lambda x: calendar.month_abbr[x])
```

```python
ue.drop(columns=['Frequency','month_int'])
```

```python
df1=ue[['State','Estimated Unemployment Rate']].groupby('State').sum().sort_values(by='Estimated Unemployment Rate',ascending=False)
```
